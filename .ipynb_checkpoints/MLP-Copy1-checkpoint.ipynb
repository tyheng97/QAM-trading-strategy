```python
import pandas as pd
import numpy as np
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Dropout, Input
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

from numpy.random import seed
seed(2021)
from tensorflow.random import set_seed
set_seed(2021)
```

### 3 Indexes

```python
data3 = pd.read_csv("./data_3_indexes.csv")

feature_columns = data3.columns[:-3]
predict_column = data3.columns[-1]

data3
```

```python
train_data_ratio = 0.9

train_data = data3[:int(train_data_ratio*len(data3))]
test_data = data3[int(train_data_ratio*len(data3)):]

print("Train data size: {}, Test data size: {}".format(len(train_data), len(test_data)))
```

#### Model definition

```python
BATCH_SIZE = 32
EPOCHS = 100
LEARNING_RATE = 0.001
```

```python
def get_model(layers=1, units=10, drop_out_rate=0):
    model = Sequential()
    model.add(Input(shape=len(feature_columns)))
    
    for layer in range(layers):
        model.add(Dense(units, activation='relu'))
        model.add(Dropout(drop_out_rate))
    
    model.add(Dense(1, activation='linear')) # output layer
    
    model.compile(optimizer=Adam(learning_rate=LEARNING_RATE),
                  loss='mean_squared_error', metrics="mean_squared_error")
    return model
```

#### Search for best parameters

```python
search_space = {
    "layers": [i for i in range(1,4)],
    "units": [i*10 for i in range(1,4)],
    "dropout_rates": [i/10 for i in range(1,4)]
}
search_space
```

```python
results = list()

for layer in search_space['layers']:
    for unit in search_space['units']:
        for dropout_rate in search_space['dropout_rates']:
            model = get_model(layer, unit, dropout_rate)
            model.fit(x=train_data[feature_columns], y=train_data[predict_column],
                      epochs=EPOCHS, verbose=0, batch_size=BATCH_SIZE,
                      validation_data=(test_data[feature_columns], test_data[predict_column]))
            preds = model.predict(test_data[feature_columns])
            mse = mean_squared_error(test_data[predict_column], preds)
            results.append((mse, layer, unit, dropout_rate))
```

```python
sorted(results, key=lambda x: x[0])
```

#### Model Training

```python
model = get_model(1, 30, 0.3)
history = model.fit(x=train_data[feature_columns], y=train_data[predict_column],
                    epochs=EPOCHS, verbose=1, batch_size=BATCH_SIZE,# callbacks=[learning_rate_decay],
                    validation_data=(test_data[feature_columns], test_data[predict_column]))
```

#### Evaluate and Plot

```python
preds = model.predict(test_data[feature_columns])
mse = mean_squared_error(test_data[predict_column], preds)
print("MSE: {}".format(mse))
```

```python
plt.plot([i for i in range(len(test_data))], test_data[predict_column], "b-")
plt.plot([i for i in range(len(test_data))], preds, "r--")
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.show()
```
